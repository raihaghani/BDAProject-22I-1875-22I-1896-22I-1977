```python
import os
from pymongo import MongoClient
from gridfs import GridFS
import csv

# Function to store audio files in MongoDB using GridFS
def store_audio_files(audio_folder):
    # Connect to MongoDB
    client = MongoClient('mongodb://localhost:27017/')
    db = client['audio_files']
    fs = GridFS(db)

    # Iterate over audio files and store them in MongoDB
    for subdir in os.listdir(audio_folder):
        subdir_path = os.path.join(audio_folder, subdir)
        if os.path.isdir(subdir_path):
            for filename in os.listdir(subdir_path):
                file_path = os.path.join(subdir_path, filename)
                if file_path.endswith('.mp3'):
                    try:
                        with open(file_path, 'rb') as audio_file:
                            fs.put(audio_file, filename=filename)
                        print(f"Stored {filename} in MongoDB")
                    except Exception as e:
                        print(f"Error storing {filename}: {e}")

# Function to store features in a CSV file
def store_features_to_csv(metadata_features, csv_file):
    # Write features to CSV file
    with open(csv_file, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=['Artist', 'Genre', 'Track', 'MFCC', 'Spectral Centroid', 'Zero-Crossing Rate'])
        writer.writeheader()
        for data in metadata_features:
            writer.writerow(data)

    print("CSV file saved successfully.")

# Path to the folder containing audio files
audio_folder = "/home/ebraheem/Documents/BDA_PROJECT/Custom_Sample"

# Path to the directory where CSV file should be stored
csv_directory = "/home/ebraheem/Documents/BDA_PROJECT/"

# Ensure the directory exists
os.makedirs(csv_directory, exist_ok=True)

# Path to the CSV file
csv_file = os.path.join(csv_directory, "audio_features.csv")

# Metadata and features data (Replace with actual data)
metadata_features = [
    {'Artist': 'Artist1', 'Genre': 'Genre1', 'Track': 'Track1', 'MFCC': [1, 2, 3], 'Spectral Centroid': 100, 'Zero-Crossing Rate': 0.5},
    {'Artist': 'Artist2', 'Genre': 'Genre2', 'Track': 'Track2', 'MFCC': [4, 5, 6], 'Spectral Centroid': 200, 'Zero-Crossing Rate': 0.6},
    # Add more data as needed
]

# Store audio files in MongoDB
store_audio_files(audio_folder)

# Store features in CSV file
store_features_to_csv(metadata_features, csv_file)
```
